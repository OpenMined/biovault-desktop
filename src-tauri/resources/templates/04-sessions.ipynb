# Beaver Tutorial 4: Session Save/Load

```python
import beaver

# Acquire the active session (provided by the test harness via env/session.json)
session = beaver.active_session()
assert session is not None, 'Session should be available'
session
```

```python
# Create some state and save it to a private path
secret_data = {'count': 1, 'owner': session.owner}
state_path = session.save()
print(f'State saved to: {state_path}')
```

```python
# Simulate kernel reset by clearing variables and reloading from disk
secret_data = None
session.load(path=state_path, overwrite=True)
assert secret_data['count'] == 1
assert secret_data['owner'] == session.owner
print('State restored:', secret_data)
```
