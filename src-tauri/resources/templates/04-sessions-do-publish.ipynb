# DO Phase 1: Publish data and save state

```python
import sys
import beaver

session = beaver.active_session()
assert session is not None, 'Session should be available'

# Try to load prior state if we crashed after publishing
try:
    session.load(overwrite=True)
    print('Loaded prior state; skipping publish phase')
    raise SystemExit(0)
except FileNotFoundError:
    print('No prior state found; publishing data')

shared_data = 7
session.remote_vars['shared_data'] = shared_data
phase = 'published'
state_path = session.save()
print(f'State saved to: {state_path}')
# Simulate kernel exit after publishing
raise SystemExit(0)
```
